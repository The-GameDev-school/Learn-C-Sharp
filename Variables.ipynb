# Variables: المتغيرات
----------

## مستوي 0
المتغيرات هي عبارة عن إحتواء لقيمة ما من نوع ثابت مثل رقم أو اسم، وقد يكون المتغير خالي من أي قيمة
المتغيرات من اسمها تتغير قيمها حيث يمكن تغيير قيمها بسهولة وبأي وقت، بشرط أن تتوافق القيمة مع نوع المتغير

يمكن التفكير في المتغيرات كأنها صناديق عليها لصاقات تشير إلى نوع الصندوق
فلو فرضنا مثلا أن لدينا صندوق دمية
فقد يحتوي الصندوق على دمية، ويمكن تغيير الدمية بسهولة، ويمكن ترك الصندوق فارغا بلا أي دمى

من أشهر أنواع المتغيرات في السي شارب
| النوع       |       الوصف    | مثال |
| :---------: | :------------- | :----------- |
| string | عبارة عن كلمة أو عدة كلمات وتستخدم علامة " لتحديد بداية ونهاية الكلام| "Ahmad", "Mohammad has eaten an apple" |
| char | عبارة عن خانة واخدة فقط، قد تكون حرفا أو رقما أو فراغا وتستخدم العلامة ' لتحديد بداية ونهاية الخانة | 'A', '3', ' ' |
| int | رقم خال تماما من الفواصل | 4, 322
| float | رقم يحتوي علي فواصل | 4.559, 1.2
| bool | هو اما قيمه صحيحه او خاطئه ولا يمكن ان ياخد اي قيمه غيرهما| true, false |

```
// here is some example code to show how to use variables
int x = 5;
float y = 3.14;
char z = 'a';
string s = "hello";
bool b = true;
```
